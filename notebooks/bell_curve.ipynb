```python
import json
import elo
import pandas as pd
from importlib import reload
from scipy.stats import norm
import matplotlib as plt
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

elo = reload(elo)
from elo import get_state, order_state, read_counter
```

```python
state = get_state()
ordered = order_state(state)

words = pd.DataFrame([{"word": k, "relevance": v["initial_rating"]} for k, v in ordered.items()])

# sns.distplot(words.relevance.values)

y = 8
words.relevance.plot.hist(bins=11, alpha=0.5, figsize=(y*1.77, y))
```

```python
top = words[:len(words)//4]

counter = read_counter()

wcount = pd.DataFrame([{"word": e[0], "counter": counter[e[0]]} for e in top.values], index=top.word.values)

plot = wcount.plot.barh(figsize=(20,60))
# top
plt.savefig("top_word_count.png")
```
